```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athleltic_sales_2020 = pd.read_csv('athletic_sales_2020.csv')
athleltic_sales_2021 = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athleltic_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
athleltic_sales_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athleltic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
athleltic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([athleltic_sales_2020, athleltic_sales_2021], ignore_index=True)

# Reset the index
combined_sales_df.reset_index(drop=True, inplace=True)

combined_sales_df.head(-1)
```

```python
# Check if any values are null.
null_values = combined_sales_df.isnull().sum()

null_values
```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'])

combined_sales_df.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_sales_df['invoice_date'].dtype)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.

# Group by 'region' , 'state' , 'city' and sum the units_sold 
grouped_df = combined_sales_df.groupby(['region' , 'state' , 'city'])['units_sold'].sum().reset_index()

# Rename the 'products_sold' column to 'Total_Products_Sold'
grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

grouped_df_sorted = grouped_df.sort_values(by='Total_Products_Sold', ascending=False)

# Show the top 5 results
grouped_df_sorted.head()


```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_df = combined_sales_df.pivot_table(
    index=['region', 'state', 'city'], # group by region, state, and city
    values='units_sold', # The column to aggregate
    aggfunc='sum' 
)

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
top_5_products_pivot = pivot_df.sort_values(by='Total_Products_Sold', ascending=False).head(5)

top_5_products_pivot.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_grouped = combined_sales_df.groupby(['region' , 'state' , 'city'])[['total_sales']].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
total_sales_grouped.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

total_sales_grouped_sorted = total_sales_grouped.sort_values(by='Total Sales', ascending=False)

# Show the top 5 results.
total_sales_grouped_sorted.head()

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = combined_sales_df.pivot_table(
    index=['region', 'state' ,'city'],
    values=['total_sales'],
    aggfunc='sum'
    )

# Rename the "total_sales" column to "Total Sales"
total_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_total_sales = total_sales_pivot.sort_values(by='Total Sales', ascending=False).head(5)

top_5_total_sales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retail_total_sales_grouped = combined_sales_df.groupby(['retailer', 'region', 'state', 'city'])[['total_sales']].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
retail_total_sales_grouped.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_retail_sales = retail_total_sales_grouped.sort_values(by='Total Sales', ascending=False).head(5)

top_5_retail_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retail_total_sales_pivot = combined_sales_df.pivot_table(
    index=['retailer', 'region', 'state' ,'city'],
    values=['total_sales'],
    aggfunc='sum'
    ).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
retail_total_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

retail_total_sales_pivot_sorted = retail_total_sales_pivot.sort_values(by='Total Sales', ascending=False)

# Show the top 5 results.
retail_total_sales_pivot.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
womens_athletic_sales_pivot.columns
```

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_athletic_sales_pivot = pd.pivot_table(womens_athletic_sales_data,
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
)
# Optional: Rename the "total_sales" column to "Total Sales"
womens_athletic_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
womens_athletic_sales_pivot.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_daily_athletic_sales = womens_athletic_sales_pivot.resample('D')['Total Sales'].sum()

# # Sort the resampled data in descending order by 'Total Sales'
daily_total_sales_sorted = womens_daily_athletic_sales.sort_values(ascending=False).head()

# # Display the result
daily_total_sales_sorted
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

# Resample the data into daily bins and get the total sales for each day
womens_weekly_athletic_sales = womens_athletic_sales_pivot.resample('W')['Total Sales'].sum()


# Sort the resampled data in descending order by 'Total Sales'
womens_weekly_athletic_sales = womens_weekly_athletic_sales.sort_values(ascending=False)

# Display the result
womens_weekly_athletic_sales.head()
```
